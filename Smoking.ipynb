```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
smoking_dummies = pd.get_dummies(data=smoking,columns=['gender', 'oral', 'tartar'], drop_first=True)
```

```python
scaler = StandardScaler()
smoking_scaled = scaler.fit_transform(smoking_dummies.drop('smoking',axis=1))
smoking_scaled
```

```python
component_analisys = PCA()
component_analisys.fit_transform(smoking_scaled)
variance = component_analisys.explained_variance_ratio_
plt.figure(figsize=(10,7))
sns.barplot(x=np.arange(len(variance)), y=variance)
plt.xlabel('Feature Number in data set')
plt.ylabel('Importance of feature')
```

```python
X= smoking_scaled
X.shape
```

```python
y = smoking['smoking'].values
```

```python
from sklearn.model_selection import train_test_split
from xgboost.sklearn import XGBClassifier
from functions.scores import ModelMetrics
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
import optuna
from sklearn.metrics import log_loss
```

```python
def tuning_xgb():
    def objective(trial):
        params = {
        "n_estimators": trial.suggest_int('n_estimators',1,100),
        "reg_alpha":trial.suggest_float('reg_alpha',0.1,0.9),
        "max_depth" : trial.suggest_int("max_depth",1,20),
        "max_leaves": trial.suggest_int('max_leaves',0,200),
        "eta" : trial.suggest_float("eta",0.001, 1.0),
        "min_child_weight" : trial.suggest_int("min_child_weight",1,20),
        "subsample" : trial.suggest_float("subsample",0.1,0.9),
        "gamma" : trial.suggest_float("gamma",0.01,0.9),
        "gpu_id": 0,
        "tree_method" : "gpu_hist",
        "objective" : "binary:logistic",
        }
        xgb = XGBClassifier(**params)
        xgb.fit(X_train,y_train)
        y_pred = xgb.predict_proba(X_test)
        error = log_loss(y_test, y_pred)
        return np.sqrt(error)
    return objective

func1 = tuning_xgb()

study_xgb = optuna.create_study(direction="minimize",study_name="XgBoostClassifier")
study_xgb.optimize(func1,n_trials=10)
print(study_xgb.best_params)
```

```python
xgb_model = XGBClassifier(**study_xgb.best_params)
eval = ModelMetrics(xgb_model, 'classification', X_train, y_train, X_test, y_test)
eval.model_performance()
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import cross_validate, KFold
```

```python
def tuning_lgm():
    def objective(trial):
        params = {
            'boosting_type':'gbdt',
            'n_estimators': trial.suggest_int('n_estimators', 1, 130),
            'objective': 'binary',
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 1), 
            'reg_alpha': trial.suggest_loguniform('reg_alpha', 1e-8, 10.0),
            'reg_lambda': trial.suggest_loguniform('reg_lambda', 1e-8, 10.0),
            'num_leaves': trial.suggest_int('num_leaves', 2, 256),
            'colsample_bytree': trial.suggest_loguniform('colsample_bytree', .5, 1.),
            'subsample': trial.suggest_loguniform('subsample', .5, 1.),
            'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
            "device":"gpu",
            "verbose":-1,
            }
        lgm = LGBMClassifier(**params,random_state=42)
        kf = KFold(n_splits=3,shuffle=True,random_state=42)
        score = cross_validate(lgm,X=X_train,y=y_train,scoring="neg_log_loss", cv=kf)
        return -1.*np.mean(score["test_score"])
    return objective

func2 = tuning_lgm()

study_lgm = optuna.create_study(direction='minimize', study_name="LGBMClassifier")
study_lgm.optimize(func2, n_trials=5)
print(study_lgm.best_params)
```

```python
lgbm_model = LGBMClassifier(**study_lgm.best_params)
lgbm_eval = ModelMetrics(lgbm_model, 'classification', X_train, y_train, X_test, y_test)
lgbm_eval.model_performance()
```
